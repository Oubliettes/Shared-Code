```python

import pandas as pd
from textblob import TextBlob
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
from flair.data import Sentence
from flair.models import TextClassifier
Flair_Classifier=TextClassifier.load('en-sentiment')
import numpy as np
```

```python
# https://www.kaggle.com/datasets/snap/amazon-fine-food-reviews. Download it here

Reviews=pd.read_csv('Reviews.csv')
```

```python
Reviews.head()
```

```python
print(Reviews.info())
```

```python
Reviews=Reviews.drop(columns=["UserId","ProfileName","HelpfulnessNumerator","HelpfulnessDenominator","Time"])
```

```python
Reviews.head()
```

```python
new_result=Reviews.iloc[ :10] #isolate the first 10 rows
```

```python
new_result
```

```python
#Activate VADER

from nltk.sentiment.vader import SentimentIntensityAnalyzer
SIA=SentimentIntensityAnalyzer()

#iterate in rows for VADER scores
score_com=[]
#score_pos=[]
#score_neg=[]
score_all=[]
for i in range(0,new_result.shape[0]):
    score=SIA.polarity_scores(new_result.iloc[i][4])
    score1=score['compound'] #prints only the compound score
    score_com.append(score1)
    
    #score2=score['pos'] #prints only the positive score
    #score_pos.append(score2)
    
    #score3=score['neg'] #prints only the negative score
    #score_neg.append(score3)
    
    score_all.append(score) # prints all scores
    
new_result["Vader (V): Compound Score"]=score_com #add the new column to DF
#new_result["V: Positive Score"]=score_pos
#new_result["V: Negative Score"]=score_neg
new_result["V: All Scores"]=score_all
```

```python
new_result
```

```python
#iterate in rows for TextBlob scores for both polarity and subjectivity scores
measurements_blob=[]

for x in range(0,new_result.shape[0]):
    QuantTextBlob=TextBlob(new_result.iloc[x][3])
    measures=QuantTextBlob.sentiment
    measurements_blob.append(measures)

new_result["TextBlob Scores"]=measurements_blob #add new column to DF

#Print only the polarity scores of TextBlob analysis only
measurements2_blob=[]

for x in range(0,new_result.shape[0]):
    QuantTextBlob2=TextBlob(new_result.iloc[x][3])
    measures2=QuantTextBlob2.sentiment.polarity #add polarity
    measurements2_blob.append(measures2)

new_result["TextBlob Polarity"]=measurements2_blob #add new column to DF
```

```python
new_result
```

```python
#Embed Flair analysis to free-text rows

def sentiment_Flair(x_analysis):
    sentence=Sentence(x_analysis)
    Flair_Classifier.predict(sentence)
    label6=sentence.labels[0]
    score7=sentence.labels[0].score
    
    if "POSITIVE" in str(label6):
        return "Positive",score7
    elif "NEGATIVE" in str(label6):
        return "Negative",score7
    else:
        return "Neutral"
new_result['Flair Analysis: Prediction & Confidence']=new_result['Text'].apply \
(lambda x_analysis:sentiment_Flair(x_analysis))

def sentiment_Flair(x_score):
    sentence3=Sentence(x_score)
    Flair_Classifier.predict(sentence3)
    label8=sentence3.labels[0].value
    
    if "POSITIVE" in str(label8):
        return 'Positive'
    elif "NEGATIVE" in str(label8):
        return 'Negative'
    else:
        return "Neutral"
new_result['Flair Prediction']=new_result['Text'].apply(lambda x_score:sentiment_Flair(x_score))


def sentiment_Flair(x_label):
    sentence2=Sentence(x_label)
    Flair_Classifier.predict(sentence2)
    score8=sentence2.labels[0].score
    
    if "POSITIVE" in str(sentence2):
        return score8
    elif "NEGATIVE" in str(sentence2):
        return score8
    else:
        return "Neutral"
new_result['Flair Confidence Score']=new_result['Text'].apply(lambda x_label:sentiment_Flair(x_label))
```
